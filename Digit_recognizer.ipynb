## Inference:

The given data is clean data. We don't have any missing values present.


### Scaling the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
scaled_train=pd.DataFrame(sc.fit_transform(newtrain),columns=newtrain.columns)
```

```python
scaled_test=pd.DataFrame(sc.fit_transform(newtest),columns=newtest.columns)
```

### Defining X and y

```python
x=scaled_train
y=train["label"]
xtest=scaled_test
```

```python
x.info()
```

### Applying Various Models

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
pred_lr=lr.fit(x,y).predict(xtest) # 91.2 accuracy
```
